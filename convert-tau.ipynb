```python
import numpy as np
import matplotlib.pyplot as plt

from theory.fitTau import xMean, xVariance, xStationaryMean, xStationaryVariance
```

```python
fN = "tau.eta0.X10"

# model params
X0        = 10
nAgents   = 1000000
eta       = 0
epsi1     = 3
epsi2     = 5
# observation params
startTime = 1e-6
endTime   = 3
obs       = 100

data = np.loadtxt(f"data/{fN}.data",delimiter=",").astype(float)
data[data==0] = 0.5
data[data==nAgents] = nAgents - 0.5
xSeries = data/nAgents

T = np.logspace(np.log10(startTime), np.log10(endTime), num=obs)

eMean = np.mean(xSeries,axis=0)
tMean = xMean(T, X0/nAgents, eta, epsi1, epsi2)
lMean = np.ones(T.shape)*xStationaryMean(eta, epsi1, epsi2)

eVar = np.var(xSeries,axis=0)
tVar = xVariance(T, X0/nAgents, eta, epsi1, epsi2)
lVar = np.ones(T.shape)*xStationaryVariance(eta, epsi1, epsi2)

plt.figure(figsize=(10,3))
plt.subplot(121)
plt.loglog()
plt.plot(T, eMean, "ro")
plt.plot(T, tMean, "k-")
plt.plot(T, lMean, "k--")
plt.subplot(122)
plt.loglog()
plt.plot(T, eVar, "ro")
plt.plot(T, tVar, "k-")
plt.plot(T, lVar, "k--")
plt.show()

out = np.vstack((T, eMean, eVar, tMean, tVar, lMean, lVar))
np.savetxt(f"data/{fN}.csv", np.log10(out).T, fmt="%.4f", delimiter=",")
```
